```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
sname=['Roshni','Krishna','Mitra','Priti']
marks=[45,56,65,43]

fig = plt.figure(figsize=(5, 5))

# creating the bar plot
plt.bar(sname, marks, color='red',
        width=0.5)

plt.xlabel("Student Name")
plt.ylabel("Marks Obtained")
plt.title("Student Wise Marks Obtained")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width,bottom=menMeans, color='b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))
plt.legend(labels=['Men', 'Women'])
plt.show()
```

```python
import matplotlib.pyplot as plt
girls_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
boys_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]
Children_grades = [14, 29, 89, 18, 9, 84, 38, 67, 34, 60]
grades_range = [10,15,20,25,30,35,40,45,50,55]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(grades_range, girls_grades, color='r')
ax.scatter(grades_range, boys_grades, color='b')
ax.scatter(grades_range, Children_grades, color='g')
ax.set_xlabel('Grades Range')
ax.set_ylabel('Grades Scored')
ax.set_title('scatter plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xlist = np.linspace(-5.0, 5.0, 120)
ylist = np.linspace(-5.0, 5.0,120)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```
